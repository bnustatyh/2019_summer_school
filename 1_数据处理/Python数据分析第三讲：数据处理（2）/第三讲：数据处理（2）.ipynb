## 原始数据

```python
import pandas as pd
```

```python
df_product = pd.read_csv("input/product.csv")
df_product
```

```python
df_category = pd.read_csv("input/category.csv")
df_category
```

## 数据合并

### 纵向合并 union/concat

```python
# 按切片获取数据
df_p1 = df_product[0:10]
df_p2 = df_product[10:20]

print("========= 用于合并的DataFrame ========= ")
display(df_p1)
display(df_p2)

df_new = pd.concat([df_p1,df_p2])
print("========= 合并后的DataFrame ========= ")
display(df_new)
```

### 横向合并

在合并操作之前，要确认DataFrame的数据粒度和实体关系（一对一、一对多&多对一、多对多）


下面的代码展示了

```python
df_merged = df_product.merge(df_category,left_on='category_id',right_on="id")
df_merged
```
