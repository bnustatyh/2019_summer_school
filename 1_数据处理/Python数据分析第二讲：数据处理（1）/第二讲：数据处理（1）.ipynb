### 选取

* 使用[ ]，loc,iloc来决定选取方式
* 使用Bool Series或切片表示来选取特定的行 


```python
# 原始数据
import pandas as pd
df = pd.read_csv("input/school_grade.csv")
df
```

#### 使用[ ]来选取DataFrame的子集

这种方式选取的DataFrame和原本的DataFrame结构一致

```python
## 通过Bool Series选取

bs = df['class']==1 # Bool Series,一个只有True或False的列
print("Bool Series")
display(bs)

df[bs] # 只会取True的部分
```

```python
# 通过切片选取，适用于想取特定行数的数据
# 切片的格式是  start_index:end_index:步长
# 索引从0开始
# 结果不包含end_index
# 步长指每次选取要跳过（步长-1）行，如果省略就是默认步长为1，不跳过

print("从第二行开始取")
display(df[1:])

print("从第二行开始取，到第7行为止，每两行选一条")
display(df[1:7:2])

```

#### 使用loc来选取
可以指定选取的列

```python
print("选取1班的同学，若干列")

# df[bs]
df2 = df.loc[bs, ['student_id','student_name','grade']]
display(df2)


print("选取1班的同学的成绩")
display(df.loc[bs,'grade'])
```

### 去除重复值


```python
data = [
    [1,'李二狗',1,'数学',70],
    [1,'李二狗',1,'数学',80],
]
df = pd.DataFrame(data,columns=['student_id','student_name','class','subject','grade'])
df
```

### 自定义处理

```python
# 添加新的字段class_name
import pandas as pd
df = pd.read_csv("input/school_grade.csv")


def f(row):
    class_number = row['class']
    row['class_name'] = str(class_number) + '班'
    return row


df = df.apply(f,axis=1)
df
```

```python
# 去除学生名字中的狗字
def f(row):
    new_name = row['student_name'].replace("狗","")
    row['student_name'] = new_name
    return row

df2 = df.apply(f,axis=1)
df2
```
